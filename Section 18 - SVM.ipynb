# Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as  plt
import pandas as pd
```

# Importing the dataset

```python
dataset = pd.read_csv('Social_Network_Ads.csv')
x = dataset.iloc[:, :-1 ].values
y = dataset.iloc[:, -1 ].values # -1 index of last column which is the dependent variable `Profit`
```

# Splitting the dataset into the Training and Testing dataset

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 0)
```

```python
print(x_train)
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

# Do not apply feature scaling on dummy variables
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

Feature scaling must be performed after splitting the dataset into training & test set in order to avoid information loss.

# Training the SVM model on the training set

```python
from sklearn.svm import SVC
classifier = SVC(kernel='linear', random_state = 0) #Linear kernel specifies a normal SVM model, random_state is set to have the same results in the course.
classifier.fit(x_train, y_train)
```

# Predticting a new result

```python
print(classifier.predict(sc.transform([[30,87000]]))) # 1st observation of test set (30 & 87000)
```

We use the transform format in order to give our predictor the correct scaled format of the data. Because our model is trained based on this assumption

# Predicting the Test Set Results

```python
y_pred = classifier.predict(x_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

# Creating a Confusion Matrix

```python
from sklearn.metrics import confusion_matrix,accuracy_score 
cm = confusion_matrix(y_test, y_pred)
accuracy = accuracy_score(y_test, y_pred)
print(cm)
print("Accuracy: " + str(accuracy))
```

# Visualising the Training set results

```python
from matplotlib.colors import ListedColormap
x_set, y_set = sc.inverse_transform(x_train), y_train
X1, X2 = np.meshgrid(np.arange(start = x_set[:, 0].min() - 10, stop = x_set[:, 0].max() + 10, step = 0.25),
                     np.arange(start = x_set[:, 1].min() - 1000, stop = x_set[:, 1].max() + 1000, step = 0.25))
plt.contourf(X1, X2, classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(x_set[y_set == j, 0], x_set[y_set == j, 1], c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Logistic Regression (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
